### Read the dataset

```python
import pandas as pd
df = pd.read_csv('50_Startups.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

### Separate X (RND, ADMIN, MKT) and Y(PROFIT)

```python
X = df[['RND', 'ADMIN', 'MKT']]
Y = df[['PROFIT']]
```

```python
X.head()
```

```python
Y.head()
```

### Built the Linear Regression Model 

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, Y)
```

```python
model.intercept_
```

```python
model.coef_
```

### Evaluate the model 

```python
# R2 score 


model.score(X,Y)
```

```python
ypred = model.predict(X)
ypred[0:5]
```

```python
Y.head()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error, r2_score, root_mean_squared_error
```

```python
def evaluate_model(X,Y):
    ypred = model.predict(X)
    mse = mean_squared_error(Y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(Y, ypred)
    mape = mean_absolute_percentage_error(Y, ypred)
    r2 = r2_score(Y, ypred)
    print(f'Mean Squared Error MSE : {mse:.2f}')
    print(f'Root Mean Absolute Error RMAE : {rmse:.2f}')
    print(f'Mean Absolute Error MAE : {mae:.2f}')
    print(f'Mean Absolute Percentage Error MAPE : {mape*100:.2f}%')
    print(f'R2 Score : {r2:.4f}')
```

```python
evaluate_model(X, Y)
```

### It is a good model as the R2 score 0.9507 > 08 

### Out of Sample Prediction 

```python
rnd = 65000
admin = 50000
mkt = 40000
xnew = [[rnd, admin, mkt]]
pred = model.predict(xnew)[0][0]
print(f'The estimated profit is : {pred:.2f}')
```
